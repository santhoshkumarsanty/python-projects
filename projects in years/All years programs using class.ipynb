```python
class year:
    
    
    def check_leap_year(self,y):
        
        if y%4==0:
            if y%100==0:
                if y%400==0:
                    print(f"{y} is a leap year")
                else:
                    print(f"{y} is not a leap year")
            else:
                print(f"{y} is a leap year")
        else:
            print(f"{y} is not a leap year")

    def valid_date(self,date,month,year):
        
        if month==1 or month==3 or month==5 or month==7 or month == 9 or month==11:
            max_day = 31

        elif month==4 or month==6 or month==8 or month==10 or month==12:
            max_day = 30

        elif year%4==0 and year%100!=0 or year%400==0:
            max_day=29

        else:
            max_day = 28

        if month<1 or month>12:
            print("This date is not valid")

        elif date<1 or date>max_day:
            print("This date is not valid")

        elif year<1700 or year>2020:
            print("This date is not valid")
        else:
            print(f"This date is valid")

    def max_date_finder(self,month,year):
        
        if month==1 or month==3 or month==5 or month==7 or month == 9 or month==11:
            max_day = 31

        elif month==4 or month==6 or month==8 or month==10 or month==12:
            max_day = 30

        elif year%4==0 and year%100!=0 or year%400==0:
            max_day=29

        else:
            max_day = 28

        return max_day

    def check_palindromic_date(self,date,month,year):
        
        max_date = self.max_date_finder(month,year)  #calling the max_date_finder its return max_date
        if date <= max_date:
            dd = str(date)
            mm = str(month)
            yy = str(year)
            if date<10:
                dd = "0"+dd
            if month <10:
                mm = "0"+mm
            check = dd+mm+yy
            reverse = check[::-1]
            if check==reverse:
                print(f"{dd}/{mm}/{yy} is a palindromic date")
            else:
                print(f"{dd}/{mm}/{yy} is not a palindromic date")

        else:
            print("This date is not valid")


    def next_palindromic_date(self,date,month,year):
        count =0
        while True:
            max_date = self.max_date_finder(month, year)
            if (date <= max_date) and (month>=1 or month<=12):
                dd = str(date)
                mm = str(month)
                yy = str(year)
                if date < 10:
                    dd = "0" + dd
                if month < 10:
                    mm = "0" + mm
                check = dd + mm + yy
                reverse = check[::-1]
                if check == reverse:
                    if count == 0:
                        print(f"{dd}/{mm}/{yy} is a palindromic date")
                        break

                    else:
                        print(f"{dd}/{mm}/{yy} is a next palindromic date")
                        break
                else:
                    count +=1
                    date += 1
                    if date > max_date:
                        date = 1
                        month +=1
                        if month > 12:
                            month = 1
                            year +=1

    
    def continues_palindromic_dates(self,s_year,e_year):
        
        for year in range(s_year,e_year+1):
            for month in range(1,13):
                max_date = self.max_date_finder(month,year)
                for date in range(1,max_date+1):
                    if (date <= max_date) and (month >= 1 or month <= 12):
                        dd = str(date)
                        mm = str(month)
                        yy = str(year)
                        if date < 10:
                            dd = "0" + dd
                        if month < 10:
                            mm = "0" + mm
                        check = dd + mm + yy
                        reverse = check[::-1]
                        if check == reverse:
                            print(f"{dd}/{mm}/{yy} is a palindromic date.")


ans = year()
ans.check_leap_year(1700)
print("----------------------------")
ans.valid_date(3,2,2020)
print("----------------------------")
ans.check_palindromic_date(21,2,2012)
print("----------------------------")
ans.next_palindromic_date(12,2,2012)
print("----------------------------")
ans.continues_palindromic_dates(1700,2026)
```
